## Mnist Dataset - Handwritten digits classification using Neural Network

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
(X_train,y_train) , (X_test,y_test) = keras.datasets.mnist.load_data()
```

Scaling the dataset

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

```python
X_train.shape
```

```python
X_train_flattened = X_train.reshape(len(X_train),28*28)
X_test_flattened = X_test.reshape(len(X_test),28*28)
```

```python
X_test_flattened.shape
```

```python
X_test_flattened[0]
```

```python
model= keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened,y_train,epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

### Sample prediction

```python
plt.matshow(X_test[0])
```

```python
y_predict = model.predict(X_test_flattened)
y_predict[0]
```

```python
np.argmax(y_predict[0])
```

```python
plt.matshow(X_test[2])
```

```python
y_predict[2]
```

```python
np.argmax(y_predict[2])
```

```python
y_predict_labels = [ np.argmax(i) for i in y_predict]
y_predict_labels[:6]
```

```python
y_test[:6]
```

```python
cm = tf.math.confusion_matrix(y_test,y_predict_labels)
cm
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

Let's try to add a Hidden Layer and check accuracy

```python
# without using flattened array
model= keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train ,y_train,epochs=5)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_predict = model.predict(X_test)
y_predict_labels = [ np.argmax(i) for i in y_predict]
cm = tf.math.confusion_matrix(labels= y_test,predictions= y_predict_labels)

plt.figure(figsize=(10,8))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
